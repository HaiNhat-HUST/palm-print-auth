```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

# Specify the correct file path with extension
img_path = 'ROI_sample.jpg'  # Replace with the actual file name and extension
img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)

# Check if the image is loaded correctly
if img is None:
    print(f"Error: The image '{img_path}' could not be loaded. Please check the file path.")
else:
    # Perform edge detection
    edges = cv.Canny(img, 5, 100)
    
    # Plot the original and edge images
    plt.figure()
    plt.subplot(121), plt.imshow(img, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(edges, cmap='gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    plt.show()





```
